```python
import pandas as pd
import pm4py 
```

```python
rtfm_log=pm4py.read_xes("./Data/road_traffic/Road_Traffic_Fine_Management_Process.xes")
```

```python
date_ours="11-27-2025_22-49-29"
date_baseline="11-28-2025_14-30-07"
```

```python
def returnPrototype(log, route):
    prot=pd.read_csv(route)
    sequence=log[log["case:concept:name"]==prot["case:concept:name"][0]]["concept:name"]
    return sequence
```

```python
prot_collected=returnPrototype(rtfm_log, "./results/Ours/rtfm/"+date_ours+"/best_predicted_instance_classcollected.csv")
```
